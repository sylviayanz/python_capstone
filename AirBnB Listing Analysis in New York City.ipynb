<H1> Data Cleaning

```python
#display all null values by columns

df.isnull().sum()

#there're a lot null values in 'last_review', 'reviews_per_month' and 'license'
#it's great portions of null values in the 3 columns, should drop them
```

```python
#drop columns with too many missing values
 
drop_columns = ['last_review','reviews_per_month', 'license']

df.drop(drop_columns, inplace = True , axis = 1) #axis=1, for columns
```

```python
#check if drop columns successfully

df.shape

#3 columns dropped
```

```python
#display 'na' values

df.isna().sum()

#'na' values in 'name' and 'host_name' columns
```

```python
#drop 'na' values

df.dropna(how = 'any', inplace = True, axis = 0) #axis = 0, for rows

#The dropna function allows we to remove rows with missing values (NaN) from a DataFrame. 
#Instead of outlining specific columns to drop, we can use the how parameter set to 'any', 
#which will drop rows containing at least one missing value across any column
#how='any', it means that a row (or column) will be dropped if it contains any missing values in it.
```

```python
#check if dropped successfully

df.isna().sum()

#all columns without 'na' values
```

```python
#calculate how many rows price = 0

count_zero_price = (df['price'] == 0).sum()
count_zero_price
```

```python
#subset rows price not equal to 0

df = df[df['price'] != 0]
```

<H2>Which Borough Has the Most AirBnB?

```python
#count all the airBnB by boroughs

neighbourhood_borough = df.groupby(['neighbourhood_group'])['host_id'].count()
neighbourhood_borough.sort_values(ascending = False)

#most hosts in Manhattan and Brooklyn
```

```python
# Create a pie chart to see the distribution of AirBnB
plt.figure(figsize=(8, 6))
plt.pie(neighbourhood_borough, labels=neighbourhood_borough.index, autopct='%1.1f%%', startangle=120, 
        colors=['salmon', 'mediumseagreen','royalblue', 'darkturquoise', 'purple'])
#startangle=120, the first slice of the pie will start 140 degrees counterclockwise from the positive x-axis.

plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Distribution of AirBnB in NYC')
plt.show()
```

```python
#group by neighbourhood_group and neighbourhood and count id numbers

neighbourhood = df.groupby(['neighbourhood_group', 'neighbourhood'])['id'].count().reset_index()
neighbourhood_sorted = neighbourhood.sort_values(by='id', ascending=False)
neighbourhood_sorted_with_rank = neighbourhood_sorted.reset_index(drop=True) #reset index column by rank
neighbourhood_sorted_with_rank.head(30) #top 30 neighbourhood of AirBnB
```

```python
import matplotlib.patches as mpatches

# Select the top 10 neighborhoods
top_10_neighbourhoods = neighbourhood_sorted_with_rank.head(10)

# Define colors for each borough
borough_colors = {
    'Manhattan': 'royalblue',
    'Brooklyn': 'mediumseagreen',
    'Queens': 'darkturquoise',
    'Bronx': 'orangered',
    'Staten Island': 'purple'
}

# Create a bar chart with colors
plt.figure(figsize=(10, 6))
bars = plt.bar(top_10_neighbourhoods['neighbourhood'], top_10_neighbourhoods['id'])

# Apply colors based on borough
for i, row in top_10_neighbourhoods.iterrows():
    borough = row['neighbourhood_group']
    color = borough_colors.get(borough)  
    bars[i].set_color(color)

plt.xlabel('Neighbourhood')
plt.ylabel('AirBnB Listing')
plt.title('Top 10 Neighbourhoods on AirBnB Listing')
plt.xticks(rotation=45, ha='right')

# Create a legend
legend_patches = [mpatches.Patch(color=color, label=borough) for borough, color in borough_colors.items()]
plt.legend(handles=legend_patches, title='Borough', loc='upper right')


plt.tight_layout()

# Show the plot
plt.show()

```

```python
# Calculate the total count of listings in all neighborhood groups
all_neighbourhoods = df['id'].count()

# Calculate the total count of listings in the top 10 neighborhood groups
top_10_neighbourhoods = neighbourhood_sorted_with_rank['id'].head(10).sum()

# Calculate the percentage
percentage_top_10_in_all = (top_10_neighbourhoods / all_neighbourhoods) * 100

print(f"Out of 223 neighborhood groups, the top 10 represent{percentage_top_10_in_all:.2f}% of all neighborhood groups.")
```

<H2>Which Room Type is Most Popular in AirBnB?

<H2>What's the Price of a Stay in AirBnB by Room Type Within Each Borough?

```python
#group by room type and price, to get median price of each room type
df.groupby('room_type')['price'].median()
```

```python
# Create boxplot to see the price by room type
room_type_price_df = df.groupby('room_type')['price'].median().reset_index()

# Draw a boxplot
sns.boxplot(data=df, x='room_type', y='price', 
            palette=[room_type_colors.get(room_type) 
                     for room_type in room_type_price_df['room_type']])

# Graph Title and Axis Label Settings
plt.title('Price Distribution by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.ylim(0, 600)

# Graph Output
plt.show()
```

```python
# Filter the dataset for Staten Island and Private room
staten_island_private_rooms = df[(df['neighbourhood_group'] == 'Staten Island') & (df['room_type'] == 'Private room')]
staten_island_private_rooms.sort_values(by = 'price', ascending = False)

#there're some extreme values, use median instead of mean price
```

```python
#create bar chart to see the price by room type in each borough

# Calculate the median price for each combination of room type and borough
average_prices = df.groupby(['neighbourhood_group', 'room_type'])['price'].median().reset_index()

# Create bar plots to show the median price for each room type within each borough
plt.figure(figsize=(12, 6))
sns.barplot(data=average_prices, x='neighbourhood_group', y='price', hue='room_type', palette=room_type_colors)

# Add labels on the bars
for p in plt.gca().patches:
    price = p.get_height()
    if not pd.isna(price):
        rounded_price = round(price) #round up price
        plt.gca().annotate(f"${rounded_price}", (p.get_x() + p.get_width() / 2, p.get_height()),
                           ha='center', va='bottom', color='black', fontsize=10)

    
plt.xlabel('Borough')
plt.ylabel('Median Price')
plt.title('Median Price by Room Type and Borough')
plt.legend(title='Room Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<div style="background-color: pink; padding: 10px"><strong>
In the New York City Airbnb dataset, the 'Hotel room' type stands out with the highest median price of 245.5 dollars per stay, making it the most expensive option. The most popular 'Entire home/apt' type follows with a median price of 174 dollars , offering good value for a private stay. 'Private room' accommodations come at a cost of around 73 dollars , while the most budget-friendly option, the 'Shared room,' has a median price of 56 dollars .         
<br><br>
When examining prices by borough, Manhattan emerges as the premium choice with the highest median prices for most room types. It boasts significantly higher prices compared to other boroughs, including Brooklyn and Queens. Interestingly, Staten Island showcases relatively consistent pricing across all room types. Median prices range from 70 to 120 dollars for 'Entire home/apt', 'Private room' and 'Shared room', and it holds the highest median price for 'Shared room' among all five boroughs.

```python
#take look of all columns
nypd_shooting.columns
```

```python
#filter columns we need
nypd_shooting = nypd_shooting[['INCIDENT_KEY', 'OCCUR_DATE','BORO','PRECINCT', 'Latitude', 'Longitude','Zip Codes']]
nypd_shooting
```

```python
#check how many null values 
nypd_shooting.isnull().sum()
```

```python
#drop 'na' values
nypd_shooting.dropna(inplace = True)
```

```python
#check if dropped
nypd_shooting
```

```python
#create a map of shooting incidents

from folium.plugins import MarkerCluster

# Create a base map centered at a specific location
shooting_map = folium.Map(location=[40.7128, -74.0060], zoom_start=10)

# Create a marker cluster
marker_cluster = MarkerCluster().add_to(shooting_map)

# Iterate through the shooting incident dataset and add markers
for index, row in nypd_shooting.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=
        f"Borough:{row['BORO']}<br>Precinct: {row['PRECINCT']}",
    ).add_to(marker_cluster)

# Display the map
display(shooting_map)
```

```python
#count shooting incidents by precinct

shooting_precinct = nypd_shooting.groupby(['PRECINCT'])['INCIDENT_KEY'].count().reset_index()
precinct_sorted = shooting_precinct.sort_values(by = 'INCIDENT_KEY',ascending = False)
precinct_sorted.reset_index(drop=True)

#there're 77 precincts in NYC, 68 of them had shooting incidents in 2023.
```

```python
#focus on top 10 AirBnB popular neighbourhoods, and see how many shooting incidents happened this year

airbnb_neighbourhood_top10 = neighbourhood_sorted_with_rank.head(10) 

#revise column names
new_column_names = {
    'neighbourhood_group': 'Borough',
    'neighbourhood': 'Neighbourhood',
    'id': 'AirBnB_Number',
    'arrest_precinct': 'Precinct',
    'shooting_number': 'Shooting_Incident'
}

# Rename the columns using the new_column_names dictionary
airbnb_neighbourhood_top10 = airbnb_neighbourhood_top10.rename(columns=new_column_names)


airbnb_neighbourhood_top10['Precinct'] = [79, 90, 14, 28, 83, 20,18, 19,71, 9]
airbnb_neighbourhood_top10['Shooting_Incident'] = [12,6 ,2,5, 6, 2, 3, 1,6, 3 ]
airbnb_neighbourhood_top10
```

```python
# Calculate the total shooting incidents
total_shooting_incidents = shooting_precinct.sum()

# Calculate the average shooting incidents per precinct 
average_shooting_per_precinct = total_shooting_incidents / 77  # Total precincts in New York City

# Print the calculated average shooting incidents per precinct
print("Average shooting incidents per precinct:", average_shooting_per_precinct)
```

```python
#Create a bar chart to see the shooting incidents in 10 neighborhoods

# Set the style of the plot
plt.style.use('seaborn')

# Calculate the average shooting incident number
avg_shooting_number = 7.7

# Create a bar chart for shooting numbers
plt.figure(figsize=(10, 6))
bars = plt.bar(airbnb_neighbourhood_top10['Neighbourhood'], airbnb_neighbourhood_top10['Shooting_Incident'], 
               color=[borough_colors.get(borough) for borough in airbnb_neighbourhood_top10['Borough']], alpha=0.7)

# Add an average line
plt.axhline(y=avg_shooting_number, color='red', linestyle='--', label='Average')

# Add labels and title
plt.xlabel('Neighbourhood')
plt.ylabel('Total Shooting Incidents')
plt.title('Shooting Incidents in Top 10 AirBnB Neighbourhoods 2023')
plt.xticks(rotation=45, ha='right')


# Create a legend with color labels and the 'Average' label
legend_labels = [plt.Line2D([0], [0], marker='o', color='w',  #the marker shape should be a circle, color white
                            label=borough, markersize=10, 
                            markerfacecolor=color) for borough, color in borough_colors.items()]
legend_labels.append(plt.Line2D([0], [0], color='red', linestyle='--', label='Average'))  


# Add the labels
plt.legend(handles=legend_labels)

# Display the plot
plt.tight_layout()
plt.show()
```
